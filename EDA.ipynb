# AI Investment Copilot - Exploratory Data Analysis (EDA)

The objective of this notebook is to analyze the stock market dataset, understand its structure, check data quality, visualize important features, and identify patterns before using it in the AI Investment Copilot project.

## Import Required Libraries

In this section, we import all the Python libraries required for data loading, analysis, visualization, and processing.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

## Load Dataset

In this section, we load the stock market dataset into a Pandas DataFrame to begin the exploratory data analysis.

```python
df = yf.download("TCS.NS", period="1y")
df.head()
```

## Display Dataset Information

This section provides basic information about the dataset, including the number of rows and columns, data types, and memory usage.

```python
df.info()
```

## Statistical Summary

This section displays the statistical summary of all numerical columns, including count, mean, standard deviation, minimum, maximum, and quartile values.

```python
df.describe()
```

## Missing Values Analysis

This section checks whether the dataset contains any missing or null values. Missing values can affect data analysis and machine learning model performance.

```python
df.isnull().sum()
```

## Duplicate Values Analysis

This section checks for duplicate records in the dataset. Duplicate data may lead to inaccurate analysis and should be identified before further processing.

```python
df.duplicated().sum()
```

## Closing Price Visualization

This graph shows the trend of the stock's closing price over time, helping to identify price movements and overall market trends.

```python
plt.figure(figsize=(12,5))
plt.plot(df["Close"])
plt.title("Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.grid(True)
plt.show()
```

## Volume Analysis

This graph shows the trading volume of the stock over time. It helps identify periods of high and low trading activity.

```python
plt.figure(figsize=(12,5))
plt.plot(df["Volume"], color="orange")
plt.title("Trading Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.grid(True)
plt.show()
```

## Correlation Analysis

This section calculates the correlation between numerical features to understand how different stock attributes are related to each other.

```python
df.corr(numeric_only=True)
```

## Conclusion

The exploratory data analysis shows that the dataset is clean and suitable for stock market analysis. The stock price changes over time, while trading volume also varies significantly. The Open, High, Low, and Close prices are strongly correlated, making them useful features for analysis and future prediction. This dataset is appropriate for building the AI Investment Copilot application.
